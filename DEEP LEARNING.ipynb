## Neural Networks

## 1.  Perceptron :
###  ( Single Layer Neural Network ) or ( Base Neural Network )

* The perceptron is a fundamental type of neural network used for binary classification tasks.
* It consists of a single layer of artificial neurons (also known as perceptrons) that take input values, apply weights, and generate an output.
* The perceptron is typically used for linearly separable data, where it learns to classify inputs into two categories based on a decision boundary.
* It finds applications in pattern recognition, image classification, and linear regression. However, the perceptron has limitations in handling complex data that is not linearly separable.
