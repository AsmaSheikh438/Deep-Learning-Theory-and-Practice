```python
import pandas as pd 
```

```python
dataset = pd.read_csv('Dataset_Cellphone.csv')
dataset.head(5)
```

```python
dataset.isnull().sum()
```

```python
input_data = dataset.iloc[:, :-1]  
output_data = dataset.iloc[:, -1]  
```

```python
# data scaling
from sklearn.preprocessing import StandardScaler
ss = StandardScaler()
ss.fit_transform(input_data)
```

```python
# convert into dataframe
input_data = pd.DataFrame(ss.fit_transform(input_data), columns=input_data.columns)
```

```python
input_data
```

```python
# now splitting of data into training and testing
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(input_data, output_data, test_size=0.2, random_state=42)
```

```python
input_data.shape
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
# Initialize the model
ann = Sequential()
```

```python
# Add layers
ann.add(Dense(8, input_dim=10, activation='relu'))  # Input layer with 8 neurons
ann.add(Dense(6, activation='relu'))               # Hidden layer with 6 neurons
ann.add(Dense(4, activation='relu'))               # Hidden layer with 4 neurons
ann.add(Dense(1, activation='sigmoid'))            # Output layer with 1 neuron
```

```python
# Compile the model
ann.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
# Display the model summary
ann.summary()
```

```python
# Train the model
ann.fit(x_train, y_train, epochs=10, batch_size=100)
```

```python
prd1 = ann.predict(x_train)
```

```python
prd = ann.predict(x_test)
```

```python
# Assuming 'prd' contains the predictions from your model
prd_data = []  # Initialize an empty list to store the results

# Iterate over the predictions
for i in prd:
    print(i[0])  # Print the predicted value
    if i[0] > 0.5:  # If the predicted value is greater than 0.5, classify as 1
        prd_data.append(1)
    else:  # Otherwise classify as 0
        prd_data.append(0)

# Now 'prd_data' contains the binary predictions (0 or 1)
print(prd_data)
```
